## Constraining total leaf area using hydraulic traits and metabolic scaling
Total water flow rate is constrained by basal metabolism on the lower end, and available precipitation as the upper bound. Hence, basal flow rate is related to water use efficiency and metabolism through:
<br>
<br>
$ \large Q_{0} = \gamma B_{0} $
<br>
Where $\gamma$ is water use efficiency (mol H2O/mol C), and $B_{0}$ is basal metabolic rate (mol C/ time). Assuming that $B_{0}$  scales according to height by a some power law, we can rewrite $Q_{0}$ in terms of height.
<br>
$ \large Q_{0} = \beta_{hQ} h^{\eta_{hQ}}$
<br> 

### Available flow rate due to precipitation
To quantify the maximal available flow rate we take into account $\gamma$, incoming precipitation $P_{inc}$ (mol H20/$m^{2} s$), and the disc area comprised by the roots $\pi r_{root}^{2}$. 
<br>

$ \large Q_{p} = \gamma \pi r_{root}^{2} P_{inc} $

<br>
### Total or realized flow rate
Total volume flow rate is given by evaporative molar flux $E$ (mol H2O/$m^{2}s$), the total area of stomatal openings $a_{f}$, and the molar mass and density of water.
<br>
<br>

$ \large Q_{e} = \frac{a_{f}E \mu_{W}}{\rho_{W}} $
<br>
<br>

$E$ can be found by solving the energy budget equation
<br>
<br>
$ \large R_{abs} - La_{g} - Ha_{j} - \lambda E a_{f} = 0 $
<br>
<br>

Following Kempes et al
<br>
$ \large E = \frac{f_{1}}{\lambda}\left[\frac{R_{abs} - f_{2}a_{f} - g_{2}a_{g}}{f_{1}a_{f} + g_{1}a_{g} + j_{1}a_{j}} \right] +\frac{f_{2}}{\lambda}$
<br>
Where $f_{1}$, $f_{2}$, $g_{1}$, $g_{2}$, and $j_{1}$ encode the effect of various environmental parameters and $R_{abs}$ is dependent on leaf dimensions, canopy geometry and incoming solar radiation. We note that $a_{g}$, $a_{f}$, and $a_{j}$ can all be written in terms of the total one-sided leaf area $a_{L}$, which can be related to LMA.
<br>
$ \large a_{L} = a_{l}n^{N} = \left<\frac{M_{l}}{LMA}\right>n^{N} $
<br>
We can assume that the average of leaf mass over LMA is well approximated by average leaf mass over the average reciprocal of LMA.
<br>
$ \large a_{L} = n^{N}\left<\frac{1}{LMA}\right>\frac{M_{L}}{n^{N}}$
<br> 
Hence, given a scaling relationship between photosynthetic biomass and total biomass, we can approximate one sided leaf area as:
<br>
<br>
$ \large a_{L} = \beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right> $
<br> 
<br>

<br>
<br>
$ \large a_{l} = \dfrac{a_{L}}{n^{N}}$
<br> 
<br>
Substituting into $a_{g}$, $a_{f}$, and $a_{j}$ we get
<br>

$ \large a_{f} = 2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>$
<br>
$ \large a_{g} = \beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>$
<br>
$ \large a_{j} = 2\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>$
<br>

$R_{abs}$ is defined as
<br>

$\large R_{abs} = \left[1 - \zeta_{c} - (1-\zeta_{s})\tau \right]P_{can}R_{inc} $
<br>

Where canpoy reflection $\zeta_{c}$, and transmission $\tau$ depend on leaf area index (LAI).
<br>

$\large LAI = \frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}} $
<br>

Substituting into $\tau$ and $\zeta_{c}$:
<br>

$ \large \tau = \exp\left[-\sqrt{\alpha}K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{\pi r_{can}^{2}}\right] $
<br>
And
<br>
$ \large \zeta_{c} = \frac{\zeta_{c}^{*} + f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]}{1 + \zeta_{c}^{*}f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]} $
<br>

Substituting into $R_{abs}$:

$ \large R_{abs} = \left[1 - \frac{\zeta_{c}^{*} + f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]}{1 + \zeta_{c}^{*}f\exp\left[-2K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{ \pi r_{can}^{2}}\right]} - (1 - \zeta_{s})\exp\left[-\sqrt{\alpha}K(\psi)\frac{\beta_{LM} M^{\eta_{LM}}\left<LMA^{-1}\right>}{\pi r_{can}^{2}}\right] \right]P_{can}R_{inc}$

Finally, we get an expression for $Q_{e}$ as a function of $\left<LMA\right>$.
<br>

$ \large Q_{e} = \frac{\mu_{W}}{\rho_{W}}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) \left\{\frac{f_{1}}{\lambda}\left[\frac{R_{abs}\left\{\left<LMA\right>\right\} - f_{2}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) + g_{2}\left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)}{f_{1}\left(2\beta_{LM}M^{\eta_{LM}}\delta_{s}a_{s}\left<LMA^{-1}\right>\right) + g_{1}\left(\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right) + j_{1}\left(2\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right>\right)} \right] +\frac{f_{2}}{\lambda}\right\} $
<br>

We can further note that $f_{1}$, $f_{2}$, and $j_{1}$ depend on leaf area through $g_{\upsilon}$ and $g_{Ha}$.

$  g_{\upsilon} = \frac{1}{g_{\upsilon L}^{-1} + g_{\upsilon a}^{-1}}$,  $ g_{Ha} = \frac{0.664 \hat{\rho}_{a}D_{H}R_{e_{a}}^{1/2}P_{r_{a}}^{1/3}}{d}$, and $  g_{\upsilon a} = \frac{0.664 \hat{\rho}_{a}D_{\nu}R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}}{d}$
<br>
Where $ d=1.62(\frac{a_{l}}{\pi})^{1/2} =1.62\left(\frac{\beta_{LM}M^{\eta_{LM}}\left<LMA^{-1}\right> }{n^{N}\pi}\right)^{1/2}$

And $R_{e_{a}} = \frac{ud}{\nu_{a}} = \hat{R}_{e_{a}}d$, with $u$ being wind speed and $\nu_{a}$ the kinematic viscosity of air.
<br>
To simplify the algebra we make the LMA dependence of $d$ explicit by defining $d = \hat{d}\left<LMA^{-1}\right>^{1/2}$. Substituting into $f_{1}$, $f_{2}$, and $j_{1}$ we get:
<br>

$\large j_{1} = c_{p}g_{Ha} = c_{p}\frac{0.664 \hat{\rho}_{a}D_{H}\hat{R}_{e_{a}}^{1/2}P_{r_{a}}^{1/3}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}} = \frac{c_{p}\hat{g}_{Ha}}{\left<LMA^{-1}\right>^{1/4}} = \frac{\hat{j}_{1}}{\left<LMA^{-1}\right>^{1/4}}$
<br>

$ \large f_{1} = \lambda g_{\nu} \frac{d e_{s}}{dT_{A}} \frac{0.664\hat{\rho}_{a}D{\nu}\hat{R}_{e_{a}}^{1/2}S_{c_{a}}^{1/3}a_{s}\delta_{s}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}a_{s}\delta_{s} + 0.664R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}z_{s}}$
<br>

$ \large f_{2} = \lambda g_{\nu} \frac{D_{\nu}}{p_{a}} \frac{0.664\hat{\rho}_{a}D{\nu}\hat{R}_{e_{a}}^{1/2}S_{c_{a}}^{1/3}a_{s}\delta_{s}}{\hat{d}^{1/2}\left<LMA^{-1}\right>^{1/4}a_{s}\delta_{s} + 0.664R_{e_{a}}^{1/2}S_{c_{a}}^{1/3}z_{s}}$





## Incorporating hydraulic constraints into assimilation rate
We can describe the gross biomass assimilation rate by multiplying average area-specific assimilation $A_{max}$ by total one-sided leaf area $a_{L}$.
<br>
\begin{equation}
    A_{can} = A_{max}a_{L}
\end{equation}

In absence of sufficient water the assimilation rate suffers a correction proportional to the water deficit. The mechanism through which water limitation decreases $A_{can}$ is taken to emerge from the regulation of stomatal behavior. In more precise terms, whenever the rate of evaporative flux exceeds the rate set by precipitation, stomatal conductance (here denoted as $g_{ua}$) decreases to compensate for the more demanding environmental conditions. The corresponding compensation calculated is accomplished by evaporative molar flux equal to the available amount of evapotranspiration.

\begin{equation}
    Q_{p} = \dfrac{a_{f}E \mu_{w}}{\rho_{w}}
\end{equation}

In other words


\begin{equation}
    \dfrac{Q_{p} \rho_{w}}{a_{f}\mu_{w}} = \frac{f_{1}}{\lambda}\left[\frac{R_{abs} - f_{2}a_{f} - g_{2}a_{g}}{f_{1}a_{f} + g_{1}a_{g} + j_{1}a_{j}} \right] +\frac{f_{2}}{\lambda}
\end{equation} 

We now note that this equality sets an upper bound to stomatal conductance through $f_{1}$ and $f_{2}$ given that:

\begin{align}
    f_{1} &= \hat{f}_{1}g_{\nu} \\
    f_{2} &= \hat{f}_{2}g_{\nu} \\
    g_{\nu} &= \dfrac{g_{ua}g_{uL}}{g_{ua} + g_{uL}}
\end{align} 


### Conductance threshold

This allows us to define a critical conductance threshold ($g_{crit}$) by solving for $g_{ua}$. After doing some algebra, we arrive at the following expression:


\begin{equation}
    g_{crit} = \dfrac{g_{uL}\left[\hat{f}_{1}(L_{1} - H_{p}a_{f}) + L_{2}\hat{f}_{2} \right]}{g_{uL}\left[\hat{f}_{1}(L_{1} - H_{p}a_{f}) + L_{2}\hat{f}_{2} \right]- L_{2}H_{p}}
\end{equation}


With $H_{p}$, $L_{1}$, and $L_{2}$ being defined as:


\begin{align}
    H_{p} &= \dfrac{Q_{p} \lambda \rho_{w}}{a_{f} \mu_{w}} \\
    L_{1} &= R_{abs} - g_{2}a_{g} \\
    L_{2} &= g_{1}a_{g} - j_{1}a_{j}
\end{align}


In order to simplify the operations we re-write the expression for $g_{crit}$ in terms of a condensed variable $\omega$.


\begin{equation}
    g_{crit} = \dfrac{g_{uL}}{g_{uL-\omega}}
\end{equation}


With $\omega = \frac{L_{2}H_{p}}{\hat{f}_{1}(L_{1} - H_{p}a_{f}) + L_{2}\hat{f}_{2}}$, unless $\frac{L_{2}H_{p}}{\hat{f}_{1}(L_{1} - H_{p}a_{f}) + L_{2}\hat{f}_{2}} \geq g_{\nu}$. In which case $\omega = g_{\nu}$. Thus, $g_{crit}$ represents a maximal constraint on stomatal conductance governed by local precipitation, leaf traits, temperature, and additional environmental factors.

### Stomatal conductance acclimation

Having defined $g_{crit}$, we must somehow account for how stomatal conductance is regulated when environmental factors push past this threshold. A straightforward method is to assume that $g_{ua}$ is downregulated to meet the available flow rate imposed by precipitation. The question then becomes: how does the downregulation of stomatal conductance impact total carbon assimilation? Empirical results show that the relationship between evaporative flow rate and assimilation rate is a proportionality modulated via water use efficiency, hence, this proportionality is already considered in the calculation for the water budget. The resulting equation governing the impact of stomatal regulation on carbon assimilation is given by:

\begin{equation}
    p(g_{ua}) = 
    \begin{cases}
         1 & \text{if $g_{ua} < g_{crit}$ }\\
         g_{ua}\omega/(g_{ua}g_{ul} - g_{ul}) & \text{otherwise}
    \end{cases}
\end{equation}
